```python
#intro to JSON
import json
```

```python
json_string = '''
            {
            "name":"bhargava",
            "age":21
            }
'''
```

```python
#json to python dict

data = json.loads(json_string)
data
```

```python
#dict to json
data['NewKey'] = True

json_format = json.dumps(data,indent=4,sort_keys=True) #indent gives inted string,sort_keys sorts the keys
print(json_format)
```

```python
#for files,
with open('file_name','r') as f:
    data = json.load(f)
    
with open('file_name','wb') as f:
    json.dump(data,f)
```

```python
payload = {'page':2,'count':25}
```

```python
r = requests.get('https://httpbin.org/get',params=payload)  #better to add parameters externally using params instead of in url manually
```

```python
r.text  
```

```python
r.url
```

```python
payload = {'username':'Bhargava','age':25}

r = requests.post('https://httpbin.org/post',data=payload)

print(r.text)
```

```python
#using json()       |we can use json module and then use json.loads() as well
r_dict = r.json()
r_dict
```

```python
#authentication 
r = requests.get('https://httpbin.org/basic-auth/bhargava/bhat',auth=('bhargava','bhat')) #values in auth=() other than the one mentioned in the url will give error 401: unauthorised access 
print(r.text)
```

```python
r = requests.get('https://httpbin.org/delay/6',timeout=3) #it will timeout after 3 sec as the server will respond after 
print(r.text)
```

```python
#handling statuscodes 
r = requests.get('https://httpbin.org/status/200')
if r.status_code != requests.codes.not_found:
    print('200')
else:
    print(requests.codes.not_found)
```

```python
from PIL import Image
from io import BytesIO
#image
headers = {'Accept':'image/png'}
r = requests.get('https://httpbin.org/image',headers=headers)
img = Image.open(BytesIO(r.content))
img
```

```python
#proxy_server
r = requests.get('https://httpbin.org/get')
print(r.text) # this will have my public IP in the response


#with proxy server, I want to send a request through a proxy server and the IP in the response will the IP of that server
proxies = {
    "http":"139.99.237.62:80",
    "https":"139.99.237.62:80"
}
r = requests.get('https://httpbin.org/get',proxies=proxies)
print(r.text) 
```
